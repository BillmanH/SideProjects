## Question: How can I know that the current trend in sessions for this week is significantly higher? 

```python
import numpy as np
import pandas as pd

#I have my personal library of matplotlib functions. Happy to share, I only left them out here for brevity.
import sys
sys.path.append("C:\Users\Bill\Documents\data_vizualization\matplotlib")
from matplotlibFunctions import *
%matplotlib inline

#Pull data 
myWD = r"C:\Users\Bill\Desktop\Tahzoo.com Redesign\GoogleAnalytics\{}"
df = pd.read_csv(myWD.format("daily_traffic.csv"))
```

```python
df.head()
```

```python
df['mean'] = df['sessions'].mean()
df['stdev'] = df['sessions'].std()
df['upperBound'] = df['mean'] + df['stdev']*2   #drawing arbitrary lines that mark the upper and lower bounds
df['lowerBound'] = df['mean'] - df['stdev']*2   #two standard deviations from the mean. 
df.head()
```

```python
bunchOfVarLine([df['sessions'].tolist(),
                df['upperBound'].tolist(),
                df['lowerBound'].tolist()],
               labels = ['sessions','upperBound','lowerBound'])
```

Looks like we had one event in the last year where traffic was abnormally high! Let's find out when that was.

```python
df[df['sessions']>df['upperBound']]
```

## Sesonality adjustment. 

```python
from datetime import datetime
df['datetime'] = df['date'].apply(lambda x: datetime.strptime(str(x),'%Y%m%d'))
df['dayOfWeek'] = df['datetime'].apply(lambda x: datetime.weekday(x))
```

```python
df_by_week = df.groupby('dayOfWeek').mean()
df_by_week
```

```python
for row in df.index.tolist():
    avg_day = df_by_week.ix[df.ix[row,'dayOfWeek'],'sessions']
    df.loc[row,"deltaWeekday"] = df.loc[row,"sessions"] - avg_day
```

```python
df.head()
```

```python
bunchOfVarLine([df['sessions'].tolist(),
                df['deltaWeekday'].tolist()],
               labels = ['sessions','deltaWeekday'])
```

## Now let's see the upper and lower bounds of the weekday delta to see which days got abnormal traffic.

```python
df['dwMean'] = df['deltaWeekday'].mean()
df['dwStdev'] = df['deltaWeekday'].std()
df['dwUpperBound'] = df['dwMean'] + df['dwStdev']*2   #copy and paste from above with some changes.
df['dwLowerBound'] = df['dwMean'] - df['dwStdev']*2 
```

```python
df.head()
```

```python
bunchOfVarLine([df['deltaWeekday'].tolist(),
                df['dwUpperBound'].tolist(),
                df['dwLowerBound'].tolist()],
               labels = ['deltaWeekday','dwUpperBound','dwLowerBound'])
```

### Lots of significant dates become apparent when adjusted for normal workweek trends. This shows both days above and days below the expected amount.

```python
df[df['deltaWeekday']>df['dwUpperBound']]
```

### There you go. ON those dates there was a very high amount of traffic on the site. Because that day was a Saturday (a day with normally low traffic), the novice observer may not have noticed the event. 

```python
df[df['deltaWeekday']<df['dwLowerBound']]
```
